```python
import numpy as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy
import random
```

```python
from scipy.special import comb
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
def randomcolor():
    rand = lambda: random.randint(0, 255)
    return '#%02X%02X%02X' % (rand(), rand(), rand()) 

def K_means(data,k):
    f1 = data[:,0]
    f2 = data[:,1]
    X = np.array(list(zip(f1, f2)))
    
    C_x = np.random.randint(0, np.max(X), size=k)
    
    C_y = np.random.randint(0, np.max(X), size=k)
    C = np.array(list(zip(C_x, C_y)), dtype=np.float32)
    print(C)
    
    plt.scatter(f1, f2, s=7)
    
    plt.savefig('no_classes.png')

    
    C_old = np.zeros(C.shape)
    
    clusters = np.zeros(len(X))
    
    error = distManh(C, C_old, None)
    
    while error != 0:
   
        for i in range(len(X)):
            distances = distManh(X[i], C)
            cluster = np.argmin(distances)
            clusters[i] = cluster
    
        C_old = deepcopy(C)
    
        for i in range(k):
            points = [X[j] for j in range(len(X)) if clusters[j] == i]
            
            if len(points) != 0: 
                C[i] = np.mean(points,axis=0)
            else: 
                C[i] = [np.random.randint(0, np.max(X), size=1),np.random.randint(0, np.max(X), size=1)]
        error = distManh(C, C_old, None)
        print(C)

    figure, ax = plt.subplots()    
    for i in range(k):
        colors = randomcolor()
        points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
        
        if len(points) != 0:
            ax.scatter(points[:, 0], points[:, 1], s=7, c=colors)
        ax.scatter(C[:, 0], C[:, 1], marker='*', s=150, c='white',edgecolors='black')
    
    print('Metrics for K_means:')
    metrics(data[:,2].astype(np.int64),clusters.astype(np.int64),'K_means')
    plt.savefig('classes_after_K_means.png')
```

```python
def distManh(a, b, ax=-1):
    return np.sum(abs(a-b),axis=ax)  

def distEuc(a, b, ax=1):
    return np.linalg.norm(a - b, axis=ax)

def metrics(clusters, classes,alg):
    tp_plus_fp = comb(np.bincount(clusters), 2).sum()
    tp_plus_fn = comb(np.bincount(classes), 2).sum()
    A = np.c_[(clusters, classes)]
    tp = sum(comb(np.bincount(A[A[:, 0] == i, 1]), 2).sum()
             for i in set(clusters))
    fp = tp_plus_fp - tp
    fn = tp_plus_fn - tp
    tn = comb(len(A), 2) - tp - fp - fn
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    rand = (tp + tn) / (tp + fp + fn + tn) 
    f1 = 2*(precision*recall)/(precision+recall) 
    fm = np.sqrt(tp/(tp+fp)*tp/(tp+fn)) 
    print("Rand Index = ", rand)
    print("F1-measure = ", f1)
    print("Fowlkes–Mallows Index  = ", fm)
    out = open('MetricsResult{0}.txt'.format(alg),'w')
    out.write('Metrics Results for %s: \nRand Index = %s \nF1-measure = %s \nFowlkes–Mallows Index  = %s \n' % (alg,rand,f1,fm))
```

```python
class som(object):
    def __init__(self, m,  d):
        
        self.u = np.random.random((m, m, d))
        self.m = m

    def plot(self, data):
        figure2, ax2= plt.subplots()
        
        ax2.scatter(data[:,0],data[:,1],s=7,c='orange')
           
        for j in range(self.m):
            for k in range(self.m):
                x, y = self.u[j, k]
                ax2.plot(x, y, 'g*', markersize=15)

       
        for j in range(1, self.m):
            for k in range(0, self.m):
                x1, y1 = self.u[j, k]
                x2, y2 = self.u[j-1, k]
                ax2.plot([x1, x2], [y1, y2], 'b')

        
        for j in range(1, self.m):
            for k in range(0, self.m):
                x1, y1 = self.u[j, k]
                x2, y2 = self.u[j, k-1]
                ax2.plot([x1, x2], [y1, y2], 'b')

        
        plt.gca().set_aspect('auto')
        plt.savefig('SOM.png')
        


    def learn(self, data, learning_rate, cooperation, Epoch):       
        for t in range(Epoch):

            if t%100 == 0:
                print('Epoch:{0}/{1},'.format(t,Epoch))
            
            f = 1 - (t/Epoch) 
            eta = learning_rate * f 
            sigma = int(np.ceil(cooperation * f)) 
            e = data[np.random.randint(data.shape[0])] 
            
            
            jwin, kwin = self.winner(e)

            
            jbeg = max(jwin - sigma, 0) 
            jend = min(jwin + sigma, self.m)
            kbeg = max(kwin - sigma, 0)
            kend = min(kwin + sigma, self.m)

            
            for j in range(jbeg, jend):
               for k in range(kbeg, kend):
                    self.u[j, k] += eta * (e - self.u[j, k])            

    def winner(self, e):
        

        mindistance = float('inf')
        jwin, kwin = 0,0
        
        for j in range(self.m):
            for k in range(self.m):
                distance = distManh(e,self.u[j,k])
                
                if distance < mindistance: 
                    mindistance = distance
                    jwin, kwin = j,k
                    
        return jwin, kwin

    def drawclusters(self,X,n):
        figure3, ax3= plt.subplots()
        cluster = AgglomerativeClustering(n_clusters=n, affinity='euclidean', linkage='ward')  
        cluster.fit_predict(X)
        ax3.scatter(X[:,0],X[:,1], c=cluster.labels_, cmap='rainbow') 
        plt.savefig('classes_after_SOM.png')
        print('Metrics for SOM: ')
        metrics(X[:,2].astype(np.int64),cluster.labels_.astype(np.int64),'SOM')
```

```python
# print('filename')
# filename = input()
filename = "/Users/M/Desktop/Нейронки/lab3/datasets/dataset_non_separate_10_20_perc_4_classes.csv"
raw = pd.read_csv(filename,sep=';',skiprows=0)
data = np.array(raw)




k=4
som_size=15
dimension = 2
learning_rate=0.1
cooperation=5
epoch = 5000


K_means(data,k)
kohonen = som(som_size, dimension)
kohonen.learn(data[:,[0,1]],learning_rate,cooperation,epoch)
kohonen.plot(data) 
kohonen.drawclusters      
kohonen.drawclusters(data,k)
```
